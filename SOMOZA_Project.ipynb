````python
# Esta clase proporciona funcionalidad relacionada con Hugging Face Transformers pipelines .
from langchain import HuggingFacePipeline

# Esta línea importa la clase AutoTokenizer de la biblioteca de transformadores.
# La clase AutoTokenizer se utiliza para cargar tokenizadores para varios modelos lingüísticos preentrenados disponibles en la clase Hugging Face model hub.
from transformers import AutoTokenizer

# Esta línea importa toda la biblioteca de transformadores, que es una popular biblioteca desarrollada por
# Hugging Face para trabajar con diversos modelos basados en transformadores en el procesamiento del lenguaje natural (PLN),
# incluyendo tanto modelos como tokenizadores.
import transformers

# Esta línea importa la biblioteca torch, que es la biblioteca principal utilizada para el aprendizaje profundo y los cálculos tensoriales en PyTorch.
import torch

# nombre del modelo que queremos utilizar
# https://huggingface.co/meta-llama/Llama-2-7b-chat-hf

model = "meta-llama/Llama-2-7b-chat-hf"

# Cargar tokenizador y modelo
tokenizer = AutoTokenizer.from_pretrained(model)

# Configurar el proceso de generación de texto
pipeline = transformers.pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 512,
                do_sample=True,
                top_k=10,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id,
                )
````

````python
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature':0})
````

````python
from langchain import PromptTemplate,  LLMChain

template = """
             Create a SQL query snippet using the below text:
              ```{text}```
              Just SQL query:
           """

prompt = PromptTemplate(template=template, input_variables=["text"])

llm_chain = LLMChain(prompt=prompt, llm=llm)

text = """  Extrae de la tabla "latam" los paises que tenga un PIB mayor a 5 millones
"""
````

````python
print(llm_chain.run(text))
````
